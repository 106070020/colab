**Example 1: generate word cloud from text**

wordcloud will analyze the frequency of every word to display. Word Cloud automatically remove stopword.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
text1=" Python is a programming language that lets you work more quickly quickly quickly and integrate integrate integrate your systems more effectively."

wc = WordCloud(max_words = 100, width = 1000, height = 1000, background_color="white")
wc.generate(text1)
plt.axis('off')
plt.imshow(wc)
plt.show()

# save wordcloud to file
# wc.to_file("wordcloud1.jpg")
```

**Example 2: generate word cloud from list of token**

Assume that you finish preprocessing (convert to lower case, remove punctuation…), tokenization, remove stopword, multiword combine, POS, NER…. After finish this process, you will have a list of token.

```python
token_list=["Python","programming","language", "work", "quickly", "quickly", "quickly", "integrate", "integrate", "integrate","systems", "effectively"]

# you need to use joint function to convert list of term to text.
text2 = ' '.join(token_list)
wc = WordCloud(max_words = 100, width = 1000, height = 1000, background_color="white")
wc.generate(text2)
plt.axis('off')
plt.imshow(wc)
plt.show()

# save wordcloud to file
# wc.to_file("wordcloud2.jpg")
```

**Example 3: Using maked wordcloud**

```python
#Download Alice mask from google drive
import gdown
url ="https://drive.google.com/uc?id=15hGqlKowHQR9y7aS2UCgrq0W7VYNjJSL"
gdown.download(url, 'alice_mask.png', quiet=False)
```

```python
from wordcloud import WordCloud
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
text3=" Python is a programming language that lets you work more quickly quickly quickly and integrate integrate integrate your systems more effectively."
alice_mask = np.array(Image.open("alice_mask.png"))#图像 == Numpy 数组 np.array

wc = WordCloud(background_color="white", max_words=100, mask=alice_mask,
               contour_width=3, contour_color='steelblue')
wc.generate(text3)
plt.axis('off')
plt.imshow(wc)
plt.show()

# save wordcloud to file
# wc.to_file("wordcloud3.jpg")
```

**Example 4: Generate wordcloud from dictionary structure**

Assuming you want to display word cloud according to term frequency, document frequency or tfidf, the input of word cloud will be dictionary structre with the term and weight for each term.


```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
dic={"Python":0.2,"programming":0.1,"language":0.1,"work":0.1, "quickly":0.5, "integrate":0.6,"systems":0.1, "effectively":0.1}

wc = WordCloud(background_color="white", max_words=1000)
wc.generate_from_frequencies(dic)#generate_from_frequencies從字典找頻率
plt.imshow(wc)
plt.axis("off")
plt.show()

# wc.to_file("wordcloud4.jpg")
```

**Example 5: Generate wordcloud for chinese**

you need to add chinese fort to display wordcloud in chinese

```python
#download chinsese font from google drive
#Download Alice mask from google drive
import gdown
url="https://drive.google.com/uc?id=1ifnY46ggoLxTo4Z9K9jUo5JFB7_y4KU_"
gdown.download(url, 'SourceHanSerifK-Light.otf', quiet=False)
```

```python
import jieba
chinese_text = "國立清華大學立校於中西文化激盪之際，不同學問的思辨，本土與國際的對話，滋養人才，學貫中西，貢獻人群，橫跨學術、藝術、文學、軍事、政治、經濟各領域，為本校傲人的傳統。學校歷經變遷，而弦歌不曾稍歇；於逆境中，自強以不息，於板蕩中，厚德以載物，是本校師生的精神與面貌，以迄於今。"
#using jieba to tokenize chinese text
token_list = jieba.cut(chinese_text)
#after tokenize, you can remove stopword or other preprocess data.
#then join data wiht"space" between term
text5= ' '.join(token_list)
print(text5)
```

```python
#generate word cloud as normal because now chinese terms are seperated in space
from wordcloud import WordCloud
import matplotlib.pyplot as plt
font_path="SourceHanSerifK-Light.otf"
wc = WordCloud(font_path=font_path, max_words = 100, width = 1000, height = 1000, background_color="white")
wc.generate(text5)
plt.axis('off')
plt.imshow(wc)
plt.show()

# save wordcloud to file
# wc.to_file("wordcloud5.jpg")
```
